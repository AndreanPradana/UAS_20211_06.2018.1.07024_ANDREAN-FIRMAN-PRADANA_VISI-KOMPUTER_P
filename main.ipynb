```python
import cv2
```

```python
cap = cv2.VideoCapture(0)
cap.set(3, 640)
cap.set(4, 480)

#Inisalisasi wajah dan mata dengan klasifikasi cascade menggunakan file xml
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
```

```python
#Variabel untuk percabangan if
first_read = True
 
ret,img = cap.read()
 
while(ret):
    ret,img = cap.read()
    #Covert gambar ke dalam grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    #penghapusan noise
    gray = cv2.bilateralFilter(gray,5,1,1)
 
    #Deteksi wajah untuk dimasukan ke dalam pengklasifikasian mata
    faces = face_cascade.detectMultiScale(gray, 1.3, 5,minSize=(200,200))
    if(len(faces)>0):
        for (x,y,w,h) in faces:
            img = cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)

            #roi_face(menandai area tertentu) yang akan dimasukkan kedalam clasifikasi mata
            roi_face = gray[y:y+h,x:x+w]
            roi_face_clr = img[y:y+h,x:x+w]
            eyes = eye_cascade.detectMultiScale(roi_face,1.3,5,minSize=(50,50))
            for (ex,ey,ew,eh) in eyes:
                cv2.rectangle(roi_face_clr,(ex,ey),(ex+ew,ey+eh),(255,0,0),2)
 
            #pengujian apakah mata terdeteksi
            if(len(eyes)>=2):
                #Apabila mata telah terdeteksi maka lanjut ke pengujian selajutnya
                if(first_read):
                    cv2.putText(img,
                    "Mata Terdeteksi Tekan M Untuk Memulai",
                    (50,50), 
                    cv2.FONT_HERSHEY_PLAIN, 1,
                    (0,255,0),2)
                else:
                    cv2.putText(img,
                    "Mata Terbuka!", (70,70),
                    cv2.FONT_HERSHEY_PLAIN, 2,
                    (255,255,255),2)
            else:
                if(first_read):
                    #Untuk memastikan apakah mata tidak terdeteksi
                    cv2.putText(img,
                    "Mata Tidak Terdeteksi!", (70,70),
                    cv2.FONT_HERSHEY_PLAIN, 2,
                    (0,0,255),2)
                else:
                    #Untuk menampilkan output apabila terdeteksi kedipan mata
                    print("Kedipan Mata Terdeteksi--------------")
                    cv2.waitKey(3000)
                    first_read=True
             
    else:
        cv2.putText(img,
        "Wajah Tidak Terdeteksi",(50,50),
        cv2.FONT_HERSHEY_PLAIN, 2,
        (0,0,255),2)
 
    #kontrol algorima
    cv2.imshow('img',img)
    a = cv2.waitKey(1)
    if(a==ord('q')):
        break
    elif(a==ord('m') and first_read):
        #untuk memulai deteksi
        first_read = False
 
cap.release()
cv2.destroyAllWindows()
```
